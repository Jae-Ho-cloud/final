```python
import pandas as pd  # pandas 라이브러리 임포트
from pyproj import Proj, transform  # pyproj에서 Proj 및 transform 임포트
import folium  # folium 라이브러리 임포트

# TM5186 좌표계와 WGS 84 좌표계의 프로젝션 정의
tm5186 = Proj(proj='tmerc', lat_0=36, lon_0=126.0, k=1, x_0=200000, y_0=500000, ellps='GRS80')  # TM5186 프로젝션 정의
wgs84 = Proj(proj='longlat', datum='WGS84')  # WGS 84 프로젝션 정의

# 변환 함수 정의
def transform_coordinates(x, y):
    """주어진 TM5186 좌표를 WGS 84 좌표로 변환하는 함수"""
    lon, lat = transform(tm5186, wgs84, x, y)  # TM5186에서 WGS 84로 좌표 변환
    return lat, lon  # 변환된 위도와 경도를 반환

# 상위 100개 데이터만 선택하여 좌표 변환 적용
df_top100 = df.head(100)  # 상위 100개 데이터 선택
df_top100['latitude'], df_top100['longitude'] = zip(*df_top100.apply(lambda row: transform_coordinates(row['gis_x_axis'], row['gis_y_axis']), axis=1))  # 변환 적용

# 변환된 결과로 새로운 데이터프레임 df2 생성
df2 = df_top100[['latitude', 'longitude']].copy()  # latitude와 longitude만 포함한 데이터프레임 생성

# 결과 출력
print(df2)  # df2의 내용 출력
```

```python
import pandas as pd  # pandas 라이브러리 임포트
from pyproj import Proj, transform  # pyproj에서 Proj 및 transform 임포트
import folium  # folium 라이브러리 임포트

# TM5186 좌표계와 WGS 84 좌표계의 프로젝션 정의
tm5186 = Proj(proj='tmerc', lat_0=37.5665, lon_0=126.978, k=1, x_0=200000, y_0=500000, ellps='GRS80')  # TM5186 프로젝션 정의
wgs84 = Proj(proj='longlat', datum='WGS84')  # WGS 84 프로젝션 정의

# 변환 함수 정의
def transform_coordinates(x, y):
    """주어진 TM5186 좌표를 WGS 84 좌표로 변환하는 함수"""
    lon, lat = transform(tm5186, wgs84, x, y)  # TM5186에서 WGS 84로 좌표 변환
    return lat, lon  # 변환된 위도와 경도를 반환

# 데이터프레임의 각 행에 대해 좌표 변환 적용
df['latitude'], df['longitude'] = zip(*df.apply(lambda row: transform_coordinates(row['gis_x_axis'], row['gis_y_axis']), axis=1))

# 변환된 결과로 새로운 데이터프레임 df2 생성
df2 = df[['latitude', 'longitude']].copy()  # latitude와 longitude만 포함한 데이터프레임 생성

# 상위 111개 데이터 선택
df2_top111 = df2.head(111)  # df2에서 상위 111개 데이터 선택

# 결과 출력
print(df2_top111)  # df2의 상위 111개 데이터 출력
```

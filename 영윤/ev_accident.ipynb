```python
# '읍면동명' 열의 고유 값과 빈도수 계산
unique_counts_county = ev_accident1['시군구명'].value_counts()

# 시각화를 위해 데이터프레임으로 변환
county_df = unique_counts_county.reset_index()
county_df.columns = ['시군구명', '빈도수']

# 바플롯 생성
plt.figure(figsize=(12, 6))  # 그래프 크기 설정
sns.barplot(data=county_df, x='시군구명', y='빈도수')

# 그래프 제목과 라벨 설정
plt.title('시군구명 별 사고 발생 빈도수', fontsize=15, fontproperties=fontprop)
plt.xlabel('시군구명', fontsize=12, fontproperties=fontprop)
plt.ylabel('빈도수', fontsize=12, fontproperties=fontprop)

# x축 라벨 회전 (긴 텍스트 가독성을 위해)
plt.xticks(rotation=45, ha='right', fontproperties=fontprop)

# 그래프 출력
plt.show()
```
